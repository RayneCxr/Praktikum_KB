### 4. Melakukan Data cleaning pada data dengan nilai null

```python
gaby.isnull().sum()
```

```python
gaby.loc[500:, 'battery_type'] = np.nan #object
gaby.loc[500:, 'weight(g)'] = np.nan #float
gaby.loc[500:, 'ram(GB)'] = np.nan #int
```

```python
gaby[['battery_type', 'weight(g)', 'ram(GB)']].isna().sum()
```

### Mengganti nilai null sesuai ketentuan

```python
median = gaby['ram(GB)'].median()
gaby['ram(GB)'].fillna(median, inplace=True)
```

```python
mean = gaby['weight(g)'].mean()
gaby['weight(g)'].fillna(mean, inplace=True)
```

```python
modus = gaby['battery_type'].mode()[0]
gaby['battery_type'].fillna(modus, inplace=True)
```

```python
gaby.head()
```
