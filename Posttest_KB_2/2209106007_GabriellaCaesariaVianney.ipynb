##### 2. Menampilkan 10 record pertama dari dataset dengan attribute bertipe angka

```python
data = {}

data['Jumlah Record'] = gaby.shape[0]
data['Jumlah Atribut'] = gaby.shape[1]
data['Nama Atribut'] = gaby.columns.tolist()
data['Jumlah Atribut Dengan Tipe Angka'] = gaby.select_dtypes(['int64', 'float64']).shape[1]

dict = list(data.keys())

tab = prtl.PrettyTable()

tab.add_column(dict[0], [data['Jumlah Record']])
tab.add_column(dict[1], [data['Jumlah Atribut']])
tab.add_column(dict[2], ['\n'.join(data['Nama Atribut'])])
tab.add_column(dict[3], [data['Jumlah Atribut Dengan Tipe Angka']])

print(tab)
```
